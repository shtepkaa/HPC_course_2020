In my case I have 3 sendings from 3 persons. Data that I got (N-size of message):

N = 10 

4.8822e-05 seconds between Person 1 send and receive message (Time per Person)

5.1814e-05 seconds between Person 0 send and receive message (Total time)

N = 100 

4.3644e-05 seconds between Person 1 send and receive message (Time per Person)

0.000150164 seconds between Person 0 send and receive message (Total time)

N = 10000 

8.5353e-05 seconds between Person 0 send and receive message (Total time)

7.3062e-05 seconds between Person 1 send and receive message (Time per Person)

N = 1000000 

0.00205566 seconds between Person 1 send and receive message (Time per Person)

0.00304831 seconds between Person 0 send and receive message (Total time)

```
import numpy as np
size = [10,100,10000,1000000]
time =[5.1814e-05,0.000149878,8.5353e-05,0.00304831]
bandwidth=np.zeros(len(time))
for i in range(len(size)):
  bandwidth[i]=(size[i]/time[i])
```

```
import matplotlib.pyplot as plt
plt.plot(time,size)
plt.xlabel("time")
plt.ylabel("size of message")
```

```
import matplotlib.pyplot as plt
plt.plot(size,bandwidth)
plt.xlabel("size of message")
plt.ylabel("bandwidth")
```
